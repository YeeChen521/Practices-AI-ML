# VGG-16
A Convolutional Neural Network (CNN) architecture is a deep learning model designed for processing structured grid-like data such as images and is used for tasks like image classification, object detection and image segmentation

VGG-16 model is a CNN architecture, It is characterized by its depth, consisting of 16 layers including 13 convolutional layers and 3 fully connected layers. These layers are organized into blocks, with each block containing multiple convolutional layers followed by a max-pooling later for downsampling
